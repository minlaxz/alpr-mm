```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imutils
```

```python
image_path = './image08.jpg'
```

```python
image_array = cv2.imread(image_path)
```

```python
plt.imshow(image_array)
```

```python
image_array_gray = cv2.cvtColor(image_array, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_array_gray)
```

```python
blur = cv2.GaussianBlur(src=image_array_gray, ksize=(3,3), sigmaX=0)
plt.imshow(blur)
```

```python
T,thresh_image_blured = cv2.threshold(src=blur,thresh=0, maxval=255, type=cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh_image_blured)
```

```python
gray = thresh_image_blured
```

```python
edged = cv2.Canny(gray, 30, 200)
```

```python
plt.imshow(edged)
```

```python
cnts = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:10]
screenCnt = None
```

```python
for c in cnts:
	# approximate the contour
	peri = cv2.arcLength(c, True)
	approx = cv2.approxPolyDP(c, 0.005 * peri, True)
	# if our approximated contour has four points, then
	# we can assume that we have found our screen
	if len(approx) == 4:
		screenCnt = approx
		break
```
